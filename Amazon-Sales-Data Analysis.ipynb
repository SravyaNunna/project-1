```python
#Importing libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#loading dataset
data = pd.read_csv(r"C:\Users\HP\Desktop\Internship\project-1\Amazon Sales data.csv.csv")
data
```

```python
#Showing the numbers of rows and columns present in the data
data.shape
```

```python
#Showing the number of columns present in the data
data.columns
```

```python
#Showing the basic information about the data
data.info()
```

```python
data['Region'] = data['Region'].astype(str)
data['Country'] = data['Country'].astype(str)
data['Item Type'] = data['Item Type'].astype(str)
data['Sales Channel'] = data['Sales Channel'].astype(str)
data['Order Priority'] = data['Order Priority'].astype(str)
```

```python
# Using describe function on dataframe for getting basic stats of numerical dataset 
data[['Units Sold', 'Unit Price',	'Unit Cost', 'Total Revenue', 'Total Cost',	'Total Profit']].describe()
```

```python
# Checking out total null value in the all the column of data
data.isnull().sum()
```

```python
#Saving the data to sales_data
sales_data=data.copy()
```

```python
# Checking out total null value in the all the column of sales_data
sales_data.describe()
```

```python
# Display total values of all country 
pd.set_option('display.max_rows', None)
sales_data['Country'].value_counts()
```

```python
sales_data.nunique()
```

```python
country_names = sales_data.Country.value_counts().index
country_val = sales_data.Country.value_counts().values
# Pie Chart for top 20 country
fig,ax = plt.subplots(figsize=(9,9))
ax.pie(country_val[:20],labels=country_names[:20],autopct='%1.2f%%')
plt.show()
```

```python
# Creating a bar chart for Total Revenue and Order Month 
plt.bar(sales_data['Order Date'], sales_data['Total Revenue'])

# Set the chart title and axis labels
plt.title('Number of Orders Purchased')
plt.xlabel('Order Date')
plt.ylabel('Total Revenue')

# Rotate the x-axis labels for better readability

# Display the chart
plt.show()
```

```python
# Plot line graph of Total Profit and Order Year
sales_data.groupby('Order Date')['Total Profit'].mean().plot()
plt.xlabel('Order Date')
plt.ylabel('Total Profit')
plt.title('Profit per date')
```

```python
# Calculating the total revenue for each group with respect to Item Type and then sorting then in descending order
revenue_by_category = sales_data.groupby('Item Type')['Total Revenue'].sum().sort_values(ascending=False)
revenue_by_category
```

```python
# Calculating the total profit for each group with respect to Item Type and then sorting then in descending order
profit_by_category = sales_data.groupby('Item Type')['Total Profit'].sum().sort_values(ascending=False)
profit_by_category
```

```python
# Calculating correlation of 'Total Revenue', 'Total Cost' and 'Total Profit' columns present in dataframe
print(sales_data[['Total Revenue', 'Total Cost', 'Total Profit']].corr())
```
